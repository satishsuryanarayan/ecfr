# Script to interact with CFR Insights API web site and plot KPIs

```python
import json
import requests
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry
```

```python
retry_strategy = Retry(
    total=10,
    backoff_factor=1
)
adapter = HTTPAdapter(max_retries=retry_strategy)
```

# Initialize the database

```python
with requests.session() as session:
    session.mount("http://", adapter)
    session.headers.update({"Content-Type": "application/json", "Accept": "application/json"})
    json_data = json.dumps({"force": True})
    response = session.post("http://127.0.0.1:8000/initialize", data=json_data)
    print(response.status_code)
```

# Populate titles and amendments from source (eCFR API)

```python
with requests.session() as session:
    session.mount("http://", adapter)
    session.headers.update({"Content-Type": "application/json", "Accept": "application/json"})
    response = session.post("http://127.0.0.1:8000/titles")
    print(response.status_code)
```

# Populate agencies and CFR references from source (eCFR API)

```python
with requests.session() as session:
    session.mount("http://", adapter)
    session.headers.update({"Content-Type": "application/json", "Accept": "application/json"})
    response = session.post("http://127.0.0.1:8000/agencies")
    print(response.status_code)
```

# Get the list of titles and amendments

```python
with requests.session() as session:
    session.mount("http://", adapter)
    session.headers.update({"Content-Type": "application/json", "Accept": "application/json"})
    response = session.get("http://127.0.0.1:8000/titles")
    print(response.status_code)
    print(json.dumps(response.json(), indent=4))
```

# Get issue dates and amendment dates by title number

```python
with requests.session() as session:
    session.mount("http://", adapter)
    session.headers.update({"Content-Type": "application/json", "Accept": "application/json"})

    response = session.get("http://127.0.0.1:8000/titles/1")
    print(response.status_code)
    print(json.dumps(response.json(), indent=4))
    
    response = session.get("http://127.0.0.1:8000/titles/amendments/issue_dates/1")
    print(response.status_code)
    print(json.dumps(response.json(), indent=4))

    response = session.get("http://127.0.0.1:8000/titles/amendments/amendment_dates/1")
    print(response.status_code)
    print(json.dumps(response.json(), indent=4))
```

# Get the list of agencies and CFR references

```python
with requests.session() as session:
    session.mount("http://", adapter)
    session.headers.update({"Content-Type": "application/json", "Accept": "application/json"})
    response = session.get("http://127.0.0.1:8000/agencies")
    print(response.status_code)
    print(json.dumps(response.json(), indent=4))
```
